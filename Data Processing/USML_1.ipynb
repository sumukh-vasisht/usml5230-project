```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=client_id,
                  client_secret=client_secret))
```

```python
master_df = pd.read_csv('drive/MyDrive/USML/data_0.csv', nrows = 1000000)
master_df.shape
```

```python
df = master_df.iloc[:1000000]
del master_df

df = df.drop('index', axis = 1)
df = df.drop('Unnamed: 0', axis = 1)
df = df.drop('Unnamed: 0.1', axis = 1)
df.rename(columns={'duration_ms':'track_duration_ms'}, inplace=True)

print(df.shape)

start = time.time()

track_uris = df['track_uri'].tolist()

x_start = 0
x_end = 70
audio_features = []
while (x_start < len(track_uris)):
  audio_features_sublist = sp.audio_features(track_uris[x_start:x_end])
  audio_features = audio_features + audio_features_sublist
  x_start = x_end
  x_end += 70
  print('Done: ', x_start)

print('TIME TAKEN: ', time.time() - start)

df['track_details'] = audio_features

enriched_df = pd.concat([df.drop(['track_details'], axis=1), df['track_details'].apply(pd.Series)], axis=1)

print(enriched_df.shape)

# print(enriched_df.head(10))

print(enriched_df.columns)

print(enriched_df.isna().sum())
enriched_df.to_csv('drive/MyDrive/USML/enriched_data_1.csv')
```
